# Import Data Season 24/25

## Import Data Season 24/25 All Competition


```python
link_data = 'https://fbref.com/en/squads/206d90db/2024-2025/all_comps/Barcelona-Stats-All-Competitions'
df_s2425_all = pd.read_html(link_data)[0]

# Drop kolom yang level 0-nya 'Per 90 Minutes'
df_s2425_all = df_s2425_all.loc[:, df_s2425_all.columns.get_level_values(0) != 'Per 90 Minutes']

df_s2425_all.head()
```

```python
# Ambil nama kolom dengan tuple kedua (level1)
df_s2425_all.columns = [col[1] for col in df_s2425_all.columns]

# Drop beberapa kolom
df_s2425_all = df_s2425_all.drop(columns=['Nation', '90s', 'xG', 'npxG', 'xAG', 'npxG+xAG', 'PrgC', 'PrgP', 'PrgR', 'Matches'])
df_s2425_all.dropna(inplace=True)

# Ubah kolom float menjadi integer
df_s2425_all[['Age', 'Min', 'Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR']] = df_s2425_all[['Age', 'Min', 'Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR']].astype(int)

# Set index dari player
df_s2425_all.set_index('Player', inplace=True)
df_s2425_all.head()
```

```python
# Tambahkan "All Competitions" di depan setiap nama kolom
df_s2425_all.columns = ["All Competitions " + col for col in df_s2425_all.columns]
df_s2425_all.head()
```

## Season 23/24
